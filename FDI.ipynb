```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('/content/FDI data.csv')

# Set the 'Sector' column as the index
df = df.set_index('Sector')

# Transpose DataFrame to have years as rows and sectors as columns
df = df.transpose()

# Extract data for the "METALLURGICAL INDUSTRIES" sector
metallurgical_data = df['METALLURGICAL INDUSTRIES']

# Calculate year-on-year change for the "METALLURGICAL INDUSTRIES" sector
yoy_change = metallurgical_data.pct_change() * 100

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))

# Plotting year-on-year change for the "METALLURGICAL INDUSTRIES" sector
yoy_change.plot(kind='bar', ax=ax, title='Year-on-Year Change for METALLURGICAL INDUSTRIES')

plt.xlabel('Year')
plt.ylabel('Year-on-Year Change (%)')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('/content/FDI data.csv')

# Set the 'Sector' column as the index
df = df.set_index('Sector')

# Calculate the sum of values across all years for each sector
sector_sum = df.sum()

# Calculate the sector-on-sector change for the sum of all years
sos_change = sector_sum.pct_change() * 100

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))

# Plotting sector-on-sector change for the sum of all years
sos_change.plot(kind='bar', ax=ax, title='Sector-on-Sector Change for the Sum of All Years')

plt.xlabel('Sector')
plt.ylabel('Sector-on-Sector Change (%)')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('/content/FDI data.csv')

# Set the 'Sector' column as the index
df = df.set_index('Sector')

# Calculate the sum across all years for each sector
sector_sum = df.sum(axis=1)

# Calculate sector-on-sector change for the sum of all years
sos_change = sector_sum.pct_change() * 100

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))

# Plotting sector-on-sector change for the sum of all years
sos_change.plot(kind='bar', ax=ax, title='Sector-on-Sector Change for the Sum of All Years')

plt.xlabel('Sector')
plt.ylabel('Sector-on-Sector Change (%)')

plt.show()
```
